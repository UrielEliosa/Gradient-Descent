```python
from datetime import datetime
import pandas
import wget

decision =input("Do you wanna download the main files? Y/N: ")

if decision == "y":
    wget.download("https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/graphing.py")
    wget.download("https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/seattleWeather_1948-2017.csv")

import graphing

data = pandas.read_csv('seattleWeather_1948-2017.csv', parse_dates=['date'])

data['year']= [(d.year + d.timetuple().tm_yday / 365.25) for d in data.date]

print("Visual check:")
graphing.scatter_2D(data,
                   label_x="year",
                   label_y="min_temperature",
                   title="Temperatures Over Time (F)")
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula= "min_temperature ~ year",data = data).fit()

intercept = model.params[0]
slope = model.params[1]

print(f"the model is: y = {slope:0.3f} * X + {intercept:0.3f}")
```

```python
class NyNodel:
    def __init__(self):
        '''
        Creates a new MyModel
        '''
        self.slope = 0
        self.intercept = 0
    def predict(self, date):
        '''
        Estimates the temperature from the date
        '''
        return date * self.slope + self.intercept
    def get_summary(self):
        '''
        Return a string that summarises the model
        '''
        return f"y= {self.slope} * x + {self.intercept}"
    
    print("Model class ready")
```
